```python
# 轉換URL並取得URL參數
import urllib.parse as urlparse
from urllib.parse import parse_qs
```

```python
MIN_PAGE = 1
MAX_PAGE = 571
```

```python
class Knowledge():
    def __init__(self, beautifulSoupResult):
        knowledge_url = 'https://kmweb.coa.gov.tw/'
        self.id = url2id(beautifulSoupResult['href'], 'id')
        self.href = beautifulSoupResult['href']
        self.text = beautifulSoupResult.get_text()
        self.fullUrl = '{}{}'.format(knowledge_url, beautifulSoupResult['href'])
```

```python
def url2id(url, targetParam):
    parsed = urlparse.urlparse(url)
    parsed_id = parse_qs(parsed.query)[targetParam]
    return parsed_id[0]
```

```python
def getKnowledgeBase(page):
    knowledge_base_url_parm = dict()
    knowledge_base_url_parm['page']=page
    knowledge_base_url_parm['display_num']=70
    knowledge_base_url = 'https://kmweb.coa.gov.tw/knowledgebase.php'
    response = requests.post(knowledge_base_url, data = knowledge_base_url_parm)
    pageSoup = BeautifulSoup(response.content, 'lxml')
    titleList = pageSoup.select('a.title')
    # 物件List
    knowledgeList = list()
    for i in titleList:
        tknowledge = Knowledge(i)
        knowledgeList.append(tknowledge)
    return knowledgeList
```

```python
def getFullKnowledgeBase(startPageNum, endPageNum):
    fullKnowledgeList = list()
    for i in range(startPageNum, endPageNum):
        tKnowledgeList = getKnowledgeBase(i)
        fullKnowledgeList = fullKnowledgeList + tKnowledgeList
    return fullKnowledgeList
```
