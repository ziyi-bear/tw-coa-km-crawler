```python
# 引入函式庫
import requests
from bs4 import BeautifulSoup
import json
```

```python
# 轉換URL並取得URL參數
import urllib.parse as urlparse
from urllib.parse import parse_qs
```

```python
theme_list_base_url = 'https://kmweb.coa.gov.tw/theme_list.php?theme=pedia&sub_theme=km'
theme_list_base_url_parm = dict()
theme_list_base_url_parm['page']=8
theme_list_base_url_parm['display_num']=70
```

```python
class ProperNoun():
    def __init__(self, beautifulSoupResult):
        base_url = 'https://kmweb.coa.gov.tw/'
        self.id = url2id(beautifulSoupResult['href'], 'id')
        self.href = beautifulSoupResult['href']
        self.text = beautifulSoupResult.get_text()
        self.fullUrl = '{}{}'.format(base_url, beautifulSoupResult['href'])
```

```python
def url2id(url, targetParam):
    parsed = urlparse.urlparse(url)
    parsed_id = parse_qs(parsed.query)[targetParam]
    return parsed_id[0]
```

```python
def getProperNoun(page):
    theme_list_base_url_parm = dict()
    theme_list_base_url_parm['page']=page
    theme_list_base_url_parm['display_num']=70
    theme_list_base_url = 'https://kmweb.coa.gov.tw/theme_list.php?theme=pedia&sub_theme=km'
    response = requests.post(theme_list_base_url, data = theme_list_base_url_parm)
    pageSoup = BeautifulSoup(response.content, 'lxml')
    titleList = pageSoup.findAll("a", class_="fw-b")
    # 物件List
    properNounList = list()
    for i in titleList:
        tProperNoun = ProperNoun(i)
        properNounList.append(tProperNoun)
    return properNounList
```

```python
def getFullProperNoun(startPageNum, endPageNum):
    fullProperNounList = list()
    for i in range(startPageNum, endPageNum):
        tProperNounList = getProperNoun(i)
        fullProperNounList = fullProperNounList + tProperNounList
    return fullProperNounList
```

```python
getFullProperNoun(1, 5)
```
